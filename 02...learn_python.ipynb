### Python Lists

Lists are used to store multiple items in a variable.

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

Lists are created using square brackets

You can create lists with mixed datatypes

lists elements are intialised inside [   ]

```python
## Empty list
def empty_list():
    my_list = []
    print("empty list")
empty_list()

def mixed_dt():
    my_list = ["ram","23","cergy","45.0","3j"]
    print(my_list)
mixed_dt()

my_list = ["tulasi", "ram", "ponaganti"] ## Global variable

def append():
    my_list.append("Lucky")
    print(my_list)
append()

def extend():
    my_list.extend(["eisti","cytech"])
    print(my_list)
extend()

def insert():
    my_list.insert(1,"cergy")
    print(my_list)
insert()

def remove():
    my_list.remove("Lucky")
    print(my_list)
remove()

def pop():
    my_list.pop()
    print(my_list)
pop()

def sort():
    my_list.sort()
    print(my_list)
sort()

def reverse():
    my_list.reverse()
    print(my_list)
reverse()

def copy():
    my_list.copy()
    print(my_list)
copy()

def length():
    print(len(my_list))
length()

def count():
    print(my_list.count('ram'))
count()

def clear():
    my_list.clear()
    print(my_list)
clear()

def join():
    list1 = ["a", "b", "c"]
    list2 = [1, 2, 3]
    list3 = list1 + list2
    print(list3)
join()
```

### Python Tuples

Tuple is a collection of Python objects much like a list. The sequence of values stored in a tuple can be of any type, and they are indexed by integers. 

Tuple elements are intialised in ( ) 

```python
## Empty tuple with no elements
def empty_tuple():
    tuple = ()
    print("empty tuple")
empty_tuple()

## Initialized tuple
def all():
    my_tuple = ('tulasi', 'ram')
    print(my_tuple)
all() 

## Tuple with mixed datatypes
def mixed_all():
    my_tuple = ('tulasi', 'ram','25','5.0','3j')
    print(my_tuple)
mixed_all() 

## Converting list into tuple
my_list = ["tulasi","ram","ponaganti"]
def conv():
    list1 = tuple(my_list)
    print(list1)
conv()

## Accssing tuple element 
def access():
    tuple = ("tulasi","ram","ponaganti")
    list2 = tuple[1]
    print(list2)
access()

def length():
    tuple = ("tulasi","ram","ponaganti")
    list3 = len(tuple)
    print(list3)
length()

def Sorted():
    x = ["tulasi","ram","ponaganti"]
    list4 = tuple(x)
    print(sorted(list4))
Sorted()


l1 = ["tulasi","ram","ponaganti"]
# printing the tuples in object directly
for ele in enumerate(l1):
    print (ele)
    print
# changing index and printing separately
for count,ele in enumerate(l1,100):
    print (count,ele)
#getting desired output from tuple
for count,ele in enumerate(l1):
    print(count)
    print(ele)

## Looping tuple elements
def iterate():
    x = ("tulasi","ram","ponaganti")
    for i in x:
        print(i)
iterate()

## checking item exist in tuple
def exist():
    x = ("tulasi","ram","ponaganti")    
    if "ram" in x:
        print("yes")
    else:
        print("no")
exist()

## Useful methods 
my_tuple = ('a','p','p','l','e',)

# len() : get the number of elements in a tuple
print(len(my_tuple))

# count(x) : Return the number of items that is equal to x
print(my_tuple.count('p'))

# index(x) : Return index of first item that is equal to x
print(my_tuple.index('l'))

# repetition
my_tuple = ('a', 'b') * 5
print(my_tuple)

# concatenation
my_tuple = (1,2,3) + (4,5,6)
print(my_tuple)

# convert list to a tuple and vice versa
my_list = ['a', 'b', 'c', 'd']
list_to_tuple = tuple(my_list)
print(list_to_tuple)

tuple_to_list = list(list_to_tuple)
print(tuple_to_list)

# convert string to tuple
string_to_tuple = tuple('Hello')
print(string_to_tuple)
```

### Python sets

In Python, Set is an unordered collection of data type that is iterable, mutable and has no duplicate elements. The order of elements in a set is undefined though it may consist of various elements.

set elements are initialised in {  }

```python
## Empty set
def empty_set():
    set1 = set()
    print("Initial blank Set: ")
    print(set1)
empty_set()

def add():
    set = {'tulasi', 'ram', 'ponaganti'}
    set.add('lucky')
    print(set)
add()   

def remove():
    set = {'tulasi', 'ram', 'ponaganti'}
    set.remove('ram')
    print(set)
remove() 
```

### Python Dictionaries

In Python, a Dictionary can be created by placing a sequence of elements within curly {} braces, separated by ‘comma’. Dictionary holds pairs of values, one being the Key and the other corresponding pair element being its Key:value. Values in a dictionary can be of any data type and can be duplicated, whereas keys can’t be repeated and must be immutable. 

```python
Dict = {1: 'tulasi', 2: 'ram', 3: 'ponaganti'}
print("\nDictionary with the use of Integer Keys: ")
print(Dict)
 
# Creating a Dictionary
# with Mixed keys
Dict = {'Name': 'ram', 1: [1, 2, 3, 4]}
print("\nDictionary with the use of Mixed Keys: ")
print(Dict)
```

### Python Arrays

An array is a collection of items stored at contiguous memory locations. The idea is to store multiple
items of the same type together.

```python
import array as arr
 
# creating an array with integer type
a = arr.array('i', [1, 2, 3])
 
# printing original array
print ("The new created array is : ", end =" ")
for i in range (0, 3):
    print (a[i], end =" ")
print()
 
# creating an array with float type
b = arr.array('d', [2.5, 3.2, 3.3])
 
# printing original array
print ("The new created array is : ", end =" ")
for i in range (0, 3):
    print (b[i], end =" ")
```

```python
import array as arr
 
# array with int type
a = arr.array('i', [1, 2, 3])
 
 
print ("Array before insertion : ", end =" ")
for i in range (0, 3):
    print (a[i], end =" ")
print()
 
# inserting array using
# insert() function
a.insert(1, 4)
 
print ("Array after insertion : ", end =" ")
for i in (a):
    print (i, end =" ")
print()
 
# array with float type
b = arr.array('d', [2.5, 3.2, 3.3])
 
print ("Array before insertion : ", end =" ")
for i in range (0, 3):
    print (b[i], end =" ")
print()
 
# adding an element using append()
b.append(4.4)
 
print ("Array after insertion : ", end =" ")
for i in (b):
    print (i, end =" ")
print()
```
