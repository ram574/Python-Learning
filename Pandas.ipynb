```python
### importing pandas package to read data into data frame
import pandas as pd 

class DataExplorationAndStatistics:
    
    train_df = pd.read_csv("/Users/tulasiram/ponagantit@cy-tech.fr - Google Drive/My Drive/Machine_learning/python learning/car data.csv") 
    
    def __init__(self,train):
        self.train = train

    ### Data Exploration
    def exploration(self):
        ### you can use display or print functions for outputs
        print(" ------- EXPLORATION ---------- ")
        display('This is to view the table --- df.head()',self.train.head())
        display('This is to view top 10 records from data table --- df.head(10)', self.train.head(10))
        display('This is to view last 10 records from data table --- df.tail(10)', self.train.tail(10))
        display('using loc you can access index from the data --- df.loc[1:10]', self.train.loc[1:10])
        display('using loc you can access index from the data --- df.loc[:,"Year"]', self.train.loc[:,"Year"])
        display('using iloc you can also access --- df.iloc[100,300]', self.train.iloc[[100,300]])
        display('using iloc you can also access --- df.iloc[0:5,0:3]', self.train.iloc[0:5,0:3])
        display('To know the shape of the data --- df.shape', self.train.shape)
        display('To know the type of the data --- type(df)', type(self.train))
        print("To know the info of the data --- df.info()")
        display(self.train.info())
        display("Acessing subset of elements --- (['column1','column2'......])", self.train[["Year","Car_Name","Selling_Price"]])
        display("Access a single value for a row/column label pair --- df.at[4,'Year']", self.train.at[4,'Year'])
        display("Dropping a column from the dataframe --- df.drop(columns=[],axis=1)", self.train.drop(['Seller_Type'],axis=1))
        display("Randomly selecting 10 rows --- df.sample(n=10)", self.train.sample(n=10))

    ### Statistics
    def statistics(self):
        print(" ----------- STATISTICS ------- ")
        display('To know the statistics of the data --- df.describe()', self.train.describe())
        display('To know the mean of vriables from the data --- df.mean()', self.train.mean())
        display('To know the correlation of variables ------ df.corr()', self.train.corr())
        display("Returns the number of non-null values in each DataFrame column ----- df.count()", self.train.count())
        display("To know the correlation of the variables ----- df.corr()", self.train.corr())
        display("Return the list of unique values in the columns ----- df.unique()", self.train['Year'].unique())
        display("Return how many unique values are present in such feature/column ----- df.nunique()", self.train['Year'].nunique())
        display("View unique values and counts ----- df['Year'].value_counts(dropna=False)", self.train['Year'].value_counts(dropna=False))
        display("Select and order top n entries ----- df.nlargest(5,'Year'))", self.train.nlargest(5,'Year'))
        display("Select and order bottom n entries ----- df.nsmallest(5,'Year')", self.train.nsmallest(5,'Year'))
        display("Query the columns of a DataFrame with a boolean expression ----- df.query()", self.train.query("(Year=='2014') and (Fuel_Type=='Petrol')"))
        display("converting the data type ---- df['column']=df['column'].astype(str)", self.train['Fuel_Type'] == self.train['Fuel_Type'].astype(str))
        display("Replace all values' ---- df['Year'].replace(1,'2022')", self.train['Year'].replace(1,'2022'))
        display("Renaming the Geography column into Nation ---- df.rename(columns={'Present_Price':'Price'})", self.train.rename(columns={"Present_Price": "Price"}))
        display("reset index of DataFrame to row numbers, moving index to columns ---- df.reset_index()", self.train.reset_index())

    ### Handling Missing Values  
    def missing_values(self):    
        print(" ------- Handling missing values ------- ")
        display("Returns the number of null values in each DataFrame column ----- df.isnull().sum()", self.train.isnull().sum())
        display("Returns the number of null values in each DataFrame column ----- df.isna()", self.train.isna())
        display("Returns the number of null values in each DataFrame column of its sum value ----- df.isna().sum()", self.train.isna().sum())
        display("Returns the number of null values in each DataFrame column ----- df.isna().sum()", self.train.isna())
        display("Returns the number of not null values in each DataFrame column ----- df.notnull()", self.train.notnull())
        display("Returns the number of not null values in each DataFrame column ----- df.notna()", self.train.notna())

    ### Data Manipulation
    def data_manipulation(self):
        print(" -------- Data Manipulation --------- ")
        display("Ascending order : Sort values of the Balance column in ascending order ----- df.sort_values('Year')", self.train.sort_values('Year'))
        display("Descending order : Sort values of the Balance column in ascending order ----- df.sort_values('Selling_Price',ascending=False)", self.train.sort_values('Selling_Price',ascending=False))
        display("Pandas Groupby function is a versatile and easy-to-use function that helps to get an overview of the data ---- df[['Car_Name','Price']].groupby('Year').mean()",self.train[['Car_Name','Year']].groupby(self.train['Selling_Price']).count())

obj = DataExplorationAndStatistics(DataExplorationAndStatistics.train_df)
obj.exploration()
obj.statistics()
obj.missing_values()
obj.data_manipulation()
```
